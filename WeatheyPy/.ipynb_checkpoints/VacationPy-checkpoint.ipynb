```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

# Generate List of Cities

```python
# Create empty lists for holding latitudes, longitudes, countries and cities

latitudes = []
longitudes = []
cities = []
countries = []
cities_country = []

# Create a set of random latitude and longitude combinations
lats = np.random.uniform(low=-90, high=90, size=1500)
lngs = np.random.uniform(low=-180, high=180, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1])
    cityname = city.city_name
    countryname = city.country_code
    city_with_country = tuple([cityname, countryname.upper()])
    if city_with_country not in cities_country:
        cities_country.append(city_with_country)
        cities.append(city_with_country[0])
        countries.append(city_with_country[1])
        latitudes.append(lat_lng[0])
        longitudes.append(lat_lng[1])

# Print the city count     
totalcities = len(cities_country)
print(f"Total cities:  {totalcities}")
```

# Perform API Calls

Perform a weather check on each city using a series of successive API calls.
Include a print log of each city as it'sbeing processed (with the city number and city name).

```python
# Create a dataframe to hold all the city data
cities_dict = {"City": cities, "Country": countries, "City Latitude":latitudes, "City Longitude": longitudes}
cities_df = pd.DataFrame(cities_dict)
cities_df.head()
```

```python
# Add columns for temperature, humidity, windspeed, date and cloudiness into the dataframe
cities_df["Max Temperature(C)"]  = ""
cities_df["Humidity(%)"]  = ""
cities_df["Wind Speed(mps)"]  = ""
cities_df["Cloud Coverage(%)"]  = ""
cities_df["CityDateTime"] = ""
cities_df.head()
```

```python
cities_df["Max Temperature(C)"] = cities_df["Max Temperature(C)"].apply(pd.to_numeric, errors='coerce')
cities_df["Humidity(%)"] = cities_df["Humidity(%)"].apply(pd.to_numeric, errors='coerce')
cities_df["Wind Speed(mps)"] = cities_df["Wind Speed(mps)"].apply(pd.to_numeric, errors='coerce')
cities_df["Cloud Coverage(%)"] = cities_df["Cloud Coverage(%)"].apply(pd.to_numeric, errors='coerce')
cities_df["CityDateTime"] = cities_df["CityDateTime"].apply(pd.to_numeric, errors='coerce')
cities_df.dtypes
```

# Perform API calls

```python
# Use OpenWeather API
# Identify Base URL
base_url = f"http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build the query_url
query_url = f"{base_url}&units={units}&appid={weather_api_key}&q="

# Set the counter for city to see which city is being processed
city_number = 1

# Iterate through each city in the city_data_df dataframe to find the temperature, humidity, windspeed, date, cloudiness
for index, row in cities_df.iterrows():
    
    currentcity = row["City"]
  
    # Add a try-except statement: If the city is not available, print the message of missing city.
    try:
        jsonresponse = requests.get(query_url + currentcity).json()
                
        # Identify the field names for the weather parameters by studying the Open Weather API 
        # We need "main.temp", "main.humidity", "wind.speed", "clouds.all", and also taking date of calculation "dt"
        
        # Update the weather data for each city into the dataframe
        cities_df.loc[index, "Max Temperature(C)"] = jsonresponse["main"]["temp_max"]
        cities_df.loc[index, "Humidity(%)"] = jsonresponse["main"]["humidity"]
        cities_df.loc[index, "Wind Speed(mps)"] = jsonresponse["wind"]["speed"]
        cities_df.loc[index, "Cloud Coverage(%)"] = jsonresponse["clouds"]["all"]
        cities_df.loc[index, "CityDateTime"] = jsonresponse["dt"]
        
        print(f"Processing record {city_number} of {totalcities}: {currentcity}")
        city_number+=1
    except:
        print("City not found. Skipping..")
        city_number+=1
     
```

## Perform Linear Regression

### Write a function that creates the linear regression plots

```python
# Create a function for linear regerssion and plotting
from scipy.stats import linregress
def linearregression (x_values, y_values, hemisphere, parameter, latmin, latmax):
    # Create boundaries for printing the line equation on teh scatter plot
  
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    
    slope = round(slope, 2)
    intercept = round(intercept, 2)
    rvalue = round(rvalue, 2)
    pvalue = round(pvalue, 2)
    stderr = round(stderr, 2)
    
    # Get regression values
    regress_values = x_values * slope + intercept
    
    # Create line equation string
    line_eq = f"y = {slope}x + {intercept}"
        
    # Create Plot
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")

    # Label plot and annotate the line equation
    plt.xlabel('Latitude')
    plt.ylabel(f"{parameter}")
    plt.title(f"Latitude Vs. {parameter} for {hemisphere} Hemisphere on {today_date}")
    
    if hemisphere == "Northern":   
        i = 40
    else:
        i = -70
        
    if parameter == "Wind Speed(mps)":
        j = 10
    else:
        j = 20
    print(i,j)  
    
    plt.annotate(line_eq, (i,j), fontsize=15,color="red") 
    
    plt.xlim(latmin, latmax)
    
    plt.savefig(f"Output/{hemisphere}Hemisphere/LatVs.{parameter}.png")
    
    print(f"******{hemisphere} Hemisphere - Latitude Vs. {parameter} *****************")
    print(regress_values)
    print(f"\n************************************************************************")
    print(f"\nEquation of Linear regression = {line_eq}")
    print(f"\nCoefficient of Coorelation = {rvalue}")
    
    return plt.show()
```

### Northern Hemisphere Linear Regression & Scatter Plots 

```python
# Northern Hemisphere Latitude Vs. Temperature
x_values = north_hem_weather_df["City Latitude"]
y_values = north_hem_weather_df["Max Temperature(C)"]
hemisphere = "Northern"
parameter = "Max Temperature (C)"
latmin = 0
latmax = 90
linearregression (x_values, y_values, hemisphere, parameter, latmin, latmax)
```
